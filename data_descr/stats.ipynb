```python
import pandas as pd
import numpy as np

dims = [2, 3, 4, 5]

for dim in dims:
    print('-------------------------------------------')
    print('Puzzle size {}x{}'.format(dim, dim))
    df = pd.read_csv('../solver/stats_{}.csv'.format(dim))
    
    sec = np.array(df['seconds'])
    conflicts = np.array(df['conflicts'])
    learned = np.array(df['learned'])
    mb = np.array(df['MB'])
    
    print('Time mean: {}'.format(sec.mean()))
    print('Time stdev: {}'.format(sec.std()))
    print('Time min: {}'.format(sec.min()))
    print('Time max: {}'.format(sec.max()))
    
    print('Conflicts mean: {}'.format(conflicts.mean()))
    print('Conflicts stdev: {}'.format(conflicts.std()))
    print('Conflicts min: {}'.format(conflicts.min()))
    print('Conflicts max: {}'.format(conflicts.max()))
    
    print('Learned mean: {}'.format(learned.mean()))
    print('Learned stdev: {}'.format(learned.std()))
    print('Learned min: {}'.format(learned.min()))
    print('Learned max: {}'.format(learned.max()))
    
    print('MB mean: {}'.format(mb.mean()))
    print('MB stdev: {}'.format(mb.std()))
    print('MB min: {}'.format(mb.min()))
    print('MB max: {}'.format(mb.max()))
    
    print('-------------------------------------------')
    print('\n\n')
```
